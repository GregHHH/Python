```python
def route(lieu_depart, lieu_fin):
    R_prime = np.copy(R)
    ending_state = mapping[lieu_fin]
    R_prime[ending_state, ending_state] = 10000
    Q = np.array(np.zeros([20,20]))
    for i in range(1000):
        etat_actuel = np.random.randint(0,20)
        actions_possibles = []
        for j in range(20):
            if R_prime[etat_actuel, j] > 0:
                actions_possibles.append(j)
        etat_suivant = np.random.choice(actions_possibles)
        TD = R_prime[etat_actuel, etat_suivant] + gamma * Q[etat_suivant, np.argmax(Q[etat_suivant,])] - Q[etat_actuel, etat_suivant]
        Q[etat_actuel, etat_suivant] = Q[etat_actuel, etat_suivant] + alpha * TD
    route = [lieu_depart]
    lieu_suivant = lieu_depart
    while (lieu_suivant != lieu_fin):
        etat_depart = mapping[lieu_depart]
        #print(etat_depart)
        etat_suivant = np.argmax(Q[etat_depart,])
        #print(Q[etat_depart,])
        #print(etat_suivant)
        lieu_suivant = mapping_inverse[etat_suivant]
        route.append(lieu_suivant)
        lieu_depart = lieu_suivant
    return route
```
